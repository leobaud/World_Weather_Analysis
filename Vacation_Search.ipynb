```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Store the CSV you saved created in part one into a DataFrame.
city_data_df = pd.read_csv("data/WeatherPy_challenge.csv")
city_data_df.head()
```

```python
# Minimun and Maximun preference
# Ask the customer to add a minimum and maximum temperature value.
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
```

```python
# Filter the dataset to find the cities that fit the criteria.
preferred_temp_df = city_data_df.loc[(city_data_df["Max Temp"] <= max_temp) & \
                                       (city_data_df["Max Temp"] >= min_temp)]
preferred_temp_df.head(10)
```

```python
# Customer rain preference
rainy_city = str(input("Do you want it to be raining? (yes/no)"))

# Selection conditions
if (rainy_city == 'yes'):
    cities_raining = preferred_temp_df[preferred_temp_df["Rain inches (past 3 hours)"] != 0]
    if (cities_raining.empty == True):
        print('Has not been raining in the last 3 hours on the cities')
    else:
        cities_raining_count = cities_raining["Rain inches (past 3 hours)"].count()
        print(f"{cities_raining_count} cities are reporting rain.")
        
elif (rainy_city == 'no'):
    cities_raining = preferred_temp_df[preferred_temp_df["Rain inches (past 3 hours)"] == 0]
    cities_raining_count = cities_raining["Rain inches (past 3 hours)"].count()
    print(f"{cities_raining_count} cities are not reporting rain.")
else:
    print('Please enter yes or no... Ask the question again')
```

```python
# Customer snow preference
snowy_city = str(input("Do you want it to be snowing? (yes/no)"))

# Selection conditions
if (snowy_city == 'yes'):
    cities_snowing = preferred_temp_df[preferred_temp_df["Snow inches (past 3 hours)"] != 0]
    if (cities_snowing.empty == True):
        print('Has not been snowing in the last 3 hours on the cities')
    else:
        cities_snowing_count = cities_snowing["Snow inches (past 3 hours)"].count()
        print(f"{cities_snowing_count} cities are reporting snow.")
        
elif (snowy_city == 'no'):
    cities_snowing = preferred_temp_df[preferred_temp_df["Snow inches (past 3 hours)"] == 0]
    cities_snowing_count = cities_snowing["Snow inches (past 3 hours)"].count()
    print(f"{cities_snowing_count} cities are not reporting snow.")
else:
    print('Please enter yes or no... Ask the question again')
```

```python
# Create data frame on rainy and snowy cities
rainy_snowy_cities_df = pd.merge(cities_raining, cities_snowing,
                                 on=['City_ID', 'City', 'Country', 'Date', 'Lat', 'Lng', 'Max Temp', 'Humidity', 'Cloudiness', 'Wind Speed', 'Current Description', 'Rain inches (past 3 hours)', 'Snow inches (past 3 hours)'])
```

```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = rainy_snowy_cities_complete[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Set the parameters to search for a hotel in Paris.
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key,
    "location": "48.8566, 2.3522"}
# Use base URL to search for hotels in Paris.
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
# Make request and get the JSON data from the search.
hotels = requests.get(base_url, params=params).json()
hotels

# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
#     hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    
# Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
# Add a heatmap of temperature for the vacation spots and marker for each city.
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp, 
             dissipating=False, max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)
# Call the figure to plot the data.
fig
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
heat_layer = gmaps.heatmap_layer(locations, weights=max_temp,dissipating=False,
             max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
